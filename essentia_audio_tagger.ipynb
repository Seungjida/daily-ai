```python
# 환경 변수 설정 (필요 시)
import os
os.environ['PATH'] = '/usr/local/cuda/bin:' + os.environ.get('PATH', '')
os.environ['LD_LIBRARY_PATH'] = '/usr/local/cuda/lib64:' + os.environ.get('LD_LIBRARY_PATH', '')
```

```python
#cuDNN 1 : drive에 저장된 경로에서 Colab의 /content/ 폴더로 파일을 복사후 압축해제
!cp /content/drive/MyDrive/cudnn-linux-x86_64-8.6.0.163_cuda11-archive.tar.xz /content/
!tar -xvf /content/cudnn-linux-x86_64-8.6.0.163_cuda11-archive.tar.xz -C /content/
```

```python
#cuDNN 2 : 기존 손상된 cuDNN 파일 삭제
!sudo rm -rf /usr/local/cuda/lib64/libcudnn*
```

```python
#cuDNN 3 : 새 cuDNN 라이브러리 설치
# lib 디렉토리에 있는 모든 cuDNN 파일을 /usr/local/cuda/lib64/로 복사
# include/ 폴더에 있는 cuDNN 헤더 파일도 복사
!sudo cp -P /content/cudnn-linux-x86_64-8.6.0.163_cuda11-archive/lib/* /usr/local/cuda/lib64/
!sudo cp -P /content/cudnn-linux-x86_64-8.6.0.163_cuda11-archive/include/* /usr/local/cuda/include/
```

```python
#cuDNN 4 : 올바른 심볼릭 링크 생성 (복사된 libcudnn.so.8.6.0을 libcudnn.so.8 및 libcudnn.so로 연결)
!sudo ln -sf /usr/local/cuda/lib64/libcudnn.so.8.6.0 /usr/local/cuda/lib64/libcudnn.so.8
!sudo ln -sf /usr/local/cuda/lib64/libcudnn.so.8.6.0 /usr/local/cuda/lib64/libcudnn.so
```

```python
#cuDNN 5 : cuDNN 파일을 시스템에서 읽을 수 있도록 권한을 설정
!sudo chmod a+r /usr/local/cuda/lib64/libcudnn*
```

```python
#cuDNN 6 : 설치된 파일 확인
!ls -lh /usr/local/cuda/lib64/libcudnn*
```
